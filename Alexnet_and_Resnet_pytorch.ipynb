```
from torchvision import models
import torch
```

```
dir(models)
```

```
alex=models.alexnet(pretrained=True)
```

```
print(alex)
```

```
from torchvision import transforms
```

```
trans=transforms.Compose([transforms.Resize(256),transforms.CenterCrop(244),transforms.ToTensor(),transforms.Normalize(mean=[0.485, 0.456, 0.406],std=[0.229, 0.224, 0.225])])
```

```
from PIL import Image
```

```
img=Image.open('dog.jpg')
```

```
img
```

```
img_t = trans(img)
```

```
img_t
```

```
batch_t=torch.unsqueeze(img_t,0)
```

```
alex.eval()
```

```
out=alex(batch_t)
```

```
print(out.shape)
```

```
with open('Imagenet_classes.txt') as f:
  classes=[line.strip() for line in f.readlines()]
```

```
_,index=torch.max(out,1)
```

```
per=torch.nn.functional.softmax(out,dim=1)[0]*100
```

```
print(classes[index[0]],per[index[0].item()])
```

```
_,ind=torch.sort(out,descending=True)
```

```
[(classes[idx], per[idx].item()) for idx in ind[0][:5]]
```

```
res=models.resnet101(pretrained=True)
```

```
res.eval()
```

```
out=res(batch_t)
```

```
print(out.shape)
```

```
_,ind=torch.sort(out,descending=True)
```

```
per=torch.nn.functional.softmax(out,dim=1)[0]*100
```

```
[(classes[idx] , per[idx].item()) for idx in ind[0][:5]]
```
